```python
with open("/data/disk/LUO/test_only/r_tridet/TriDet/data/cataract/data_1102.json","r") as f:
    dataset = json.load(f)
```

```python
dataset['train01.mp4']["annotation"]
```

```python
reversed_dataset = copy.deepcopy(dataset)
```

```python
for k,v in dataset.items():
    if "test" in k:
        continue
    index = int(k.split("train")[-1].split(".")[0])+25
    name = "train"+str(index)+".mp4"
    reversed_dataset[name] = {}
    reversed_dataset[name]["last_time"] = v["last_time"]
    end_time = v["last_time"]
    reversed_dataset[name]["annotation"] = []
    for item in v["annotation"]:
        
        temp = {
            "start":end_time-item["end"],
            "end":end_time-item["start"],
            'duration':item["duration"],
            'label': item["label"],
            "subset":"training",
            'time_till_now': end_time-item["time_till_now"]
        }
        reversed_dataset[name]["annotation"].append(temp)
    reversed_dataset[name]["annotation"].reverse()
```

```python
def mask(feats,mask_num,mask_length):
    #give random masks to the original features
    len_feats = feats.shape[0]
    dim_feats = feats.shape[1]
    start_idx_list = [i for i in range(len_feats-mask_length-1)]
    for mask in range(mask_num):
        start_idx = np.random.choice(start_idx_list)
        feats[start_idx:start_idx+mask_length,:] = feats[start_idx:start_idx+mask_length,:]+np.random.randn(mask_length,dim_feats)

    return feats
```
